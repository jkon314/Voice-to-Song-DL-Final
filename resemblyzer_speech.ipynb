```python
# import necessary libraries
from resemblyzer import preprocess_wav, VoiceEncoder
from  pathlib import Path 
from tqdm import tqdm 
import numpy as np 
from IPython.display import Audio 
from itertools import groupby
import heapq
import torchaudio as ta
from os import listdir
import os
from os.path import isfile, join, basename
from  pathlib import Path 
import numpy as np
import torch
```

```python
dir_path = "C:/Users/baseb/.cache/huggingface/datasets/downloads/extracted"
audio_files = [join(dir_path, f) for f in listdir(dir_path) if "mp3" in f and isfile(join(dir_path, f))]

resemblyzer_vectors = []

failed_files = []
pctCounter = 1
for f in audio_files:
    os.system('cls')
    print(f'{pctCounter/len(audio_files)*100}% complete')
    pctCounter+=1
    try:
        
        #print(vocals[0].shape)
        
        #print(mono.shape)

        #print(f)
        wav = preprocess_wav(f)
        print(wav)

        #Get style encoding
        encoder = VoiceEncoder()
        embed = encoder.embed_utterance(wav)
        np.set_printoptions(precision=3, suppress=True)

        #Get Spectrogram
       

        #Add spec and style embedding to appropriate arrays
        resemblyzer_vectors.append(embed)
        
    except:
        print('failed :(')
        failed_files.append(f)

resemblyzer_vectors = np.array(resemblyzer_vectors)

print(resemblyzer_vectors.shape)
print("Number of Failed Files: ")
print(len(failed_files))
```
