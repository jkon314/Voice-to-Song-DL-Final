```python
# Create these folders manually
vocals_dir = join(output_dir, "vocals")
instrumentals_dir = join(output_dir, "instrumentals")
vocal_chunks_dir = join(output_dir, "vocals_chunks")
instrumentals_chunks_dir = join(output_dir, "instrumentals_chunks")

```

```python
import shutil

output_files = [join(output_dir, f) for f in listdir(output_dir) if isfile(join(output_dir, f))]

for f in output_files:
    if "Instrumental" in f:
        shutil.move(f, instrumentals_dir)
    elif "Vocals" in f:
        shutil.move(f, join(output_dir, vocals_dir))

```

```python
from pydub import AudioSegment
# t1 = t1 * 1000 #Works in milliseconds
# t2 = t2 * 1000

directories_to_chunk = [vocals_dir, instrumentals_dir]
output_dir = [vocal_chunks_dir, instrumentals_chunks_dir]

for k in range(len(directories_to_chunk)):
    dir_path = directories_to_chunk[k]
    audio_files = [join(dir_path, f) for f in listdir(dir_path) if "mp3" in f and isfile(join(dir_path, f))]
    print(audio_files)

    raw_split_dir = output_dir[k]
    total_num_chunks = 0
    for file in audio_files:
        print("File Name: ", file)
        chunk_size_sec = 10
        try:
            audio = AudioSegment.from_mp3(file)
            chunks = int(len(audio) / 1000 / chunk_size_sec)
            total_num_chunks += chunks
            print("Number of Chunks: ", chunks, "\n")
            for j in range(1, chunks - 1):
                t1 = j * 1000 * chunk_size_sec
                t2 = (j * 1000 * chunk_size_sec) + (chunk_size_sec * 1000)
                newAudio = audio[t1: t2]
                #print('halfway')
                file_name = basename(file).split(".mp3")[0]
                print('final stretch')
                output_file = join(raw_split_dir, "{}-chunk-{}.mp3".format(file_name, j))
                print(output_file)
                newAudio.export(output_file, format="mp3") #Exports to a wav file in the current path.
                print('SUCCESS')
        except:
            print('failed to export')
            pass

    print("Total Chunks: ", str(total_num_chunks))
```
